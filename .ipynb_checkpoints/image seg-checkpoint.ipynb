# K-means
1. assume random k points as cluster centres.
2. assign each data point to the closest cluster centre.
3. update location of cluster point to mean of all data points assigned to it.
4. repeat 2 and 3 until convergence.

```python
import cv2 
import matplotlib.pyplot as plt
import numpy as np
```

```python
img=cv2.imread('img.jpeg')
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
print(img.shape)
plt.imshow(img)
plt.show()
```

```python
# flattening pixels
allpixels=img.reshape((185*273,3))
print(allpixels.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
colors=3
km=KMeans(n_clusters=colors)
km.fit(allpixels)
```

```python
centers=km.cluster_centers_
```

```python
centers=np.array(centers,dtype='uint8')
```

```python
print(centers)
```

```python
i=1
plt.figure(0,figsize=(5,2))
colors=[]

for j in centers:
    plt.subplot(1,5,i)
#     plt.axis("off")
    i+=1
    colors.append(j)
    
#     color swatch
    a=np.zeros((200,200,3),dtype='uint8')
    a[:,:,:]=j
    plt.imshow(a)
plt.show()
```

# Segmenting our image

```python
newimg=np.zeros((185*273,3),dtype='uint8')
print(newimg.shape)
```

```python
km.labels_
```

```python
for ix in range(newimg.shape[0]):
    newimg[ix]=colors[km.labels_[ix]]
newimg=newimg.reshape((img.shape))
img2=cv2.imread('img.jpeg')
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.xlabel("Original")
plt.show()
plt.imshow(newimg)
plt.xlabel("Tranformed image")
plt.show()
```
